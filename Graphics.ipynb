```python
import matplotlib.pyplot as plt
import numpy as np

# Данные
n_values = [10**4 * i for i in range(1, 11)]  # Количество точек (n)
heap_sort_times_mode1 = [3, 7, 11, 16, 20, 25, 33, 36, 42, 528]
quick_sort_times_mode1 = [3, 11, 24, 40, 60, 88, 115, 146, 226, 40554]

heap_sort_times_mode2 = [2, 6, 9, 13, 19, 21, 25, 31, 35, 489]
quick_sort_times_mode2 = [9, 33, 58, 103, 163, 231, 315, 375, 467, 69192]

# Настройка графика
plt.figure(figsize=(14, 7))

# Режим 1
plt.subplot(1, 2, 1)
plt.plot(n_values, heap_sort_times_mode1, label="HeapSort (Mode 1)", color="blue", marker="o")
plt.plot(n_values, quick_sort_times_mode1, label="QuickSort (Mode 1)", color="orange", marker="x")
plt.yscale("log")  # Логарифмическая шкала для оси Y
plt.title("Mode 1: Random Points Inside Rectangle", fontsize=14)
plt.xlabel("Number of Points (n)", fontsize=12)
plt.ylabel("Execution Time (s) (log scale)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Режим 2
plt.subplot(1, 2, 2)
plt.plot(n_values, heap_sort_times_mode2, label="HeapSort (Mode 2)", color="green", marker="o")
plt.plot(n_values, quick_sort_times_mode2, label="QuickSort (Mode 2)", color="red", marker="x")
plt.yscale("log")  # Логарифмическая шкала для оси Y
plt.title("Mode 2: Random Points on Rectangle Border", fontsize=14)
plt.xlabel("Number of Points (n)", fontsize=12)
plt.ylabel("Execution Time (s) (log scale)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Отображение графиков
plt.tight_layout()
plt.show()
```
